### 이미지는 고정된 값인데 왜 확률분포를 사용할까?
픽셀 하나하나를 독립적인 확률변수로 두면 10x10 회색조 이미지를 표현하기 위해서는 255x10x10 만큼의 경우의 수가 필요하게 된다 (== 10x10x255 차원). 일반적인 이미지는 이거보다 사이즈가 크니까 200x200, 800x1860 같은 더 큰 사이즈의 이미지들은 차원이 훨씬 더 커진다.

이미지가 어떤 의미를 가진다는 것은 사람이 봤을 때 의미를 부여할 수 있는 어떤 형태가 존재한다는 것이고 이러한 이미지가 차원공간 전체에서 균일하게 분포한다면 랜덤 노이즈 이미지를 여러번 샘플링했을 때 의미있는 형태를 보이는 이미지가 나올 것! 

실제로 실험해보면 노이즈 이미지만 계속 나오는 것을 확인할 수 있고 의미있는 이미지들은 저 큰 차원공간에서 균일하지 않는 어떤 분포를 따르는 위치에 있을것이라고 할 수 있다.

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def random_image_generator():
    img = np.random.randint(0, 256, (100,100,3))
    return img

tmp = random_image_generator()
tmp
```

```python
imgs = [random_image_generator() for _ in range(100)]
```

```python
fig, axs = plt.subplots(10,10, figsize = (5,5))
for ax, img in zip(axs.flat, imgs):
    ax.imshow(img)
    ax.axis('off')  

plt.tight_layout()
plt.show()
```

```python
def gaussian_distributed_img():
    img = np.random.normal(125, 25, (100,100,3))
    img = np.clip(img, 0, 256)
    return img.astype(np.uint8)
```

```python
g_imgs = [gaussian_distributed_img() for _ in range(100)]

fig, axs = plt.subplots(10,10, figsize = (5,5))
for ax, img in zip(axs.flat, g_imgs):
    ax.imshow(img)
    ax.axis('off')  

plt.tight_layout()
plt.show()
```
